```python
import pandas as pd

notes_df=pd.read_csv("college_notes.csv")
print("Dataset")
print(f'shape{notes_df.shape}')
print(f'columns: {notes_df.columns.tolist}')
notes_df.head()
```

```python
all_documents=notes_df['content'].tolist()
all_ids=notes_df['note_id'].tolist()
all_metadatas=[{"subject":subject,"topic":topic} for subject,topic in zip(notes_df['subject'],notes_df['topic'])]


print(f'Total documents:{len(all_documents)}')
print("First Document")
print(f'ID:{all_ids[0]}')
print(f'Subject:{all_metadatas[0]["subject"]}')
print(f'Topic:{all_metadatas[0]["topic"]}')
print(f'Content:{all_documents[0][:100]}...')
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

query_keyword = "ETL"

model = SentenceTransformer("all-MiniLM-L6-v2")

query_embedding = model.encode(query_keyword)

results = []

for i, doc in enumerate(all_documents):

    doc_embedding = model.encode(doc)

    score = np.dot(query_embedding, doc_embedding) / (
        np.linalg.norm(query_embedding) *
        np.linalg.norm(doc_embedding)
    )

    results.append((score, i))

results.sort(reverse=True)

print(f"\nTop Matching Documents for: {query_keyword}")
print("=" * 60)

for score, i in results[:5]:

    print(f"\nRank")
    print(f'ID: {all_ids[i]}')
    print(f'Subject: {all_metadatas[i]["subject"]}')
    print(f'Topic: {all_metadatas[i]["topic"]}')
    print(f'Score: {score:.4f}')
    print(f'Content: {all_documents[i][:100]}...')
    print("=" * 60)
```
